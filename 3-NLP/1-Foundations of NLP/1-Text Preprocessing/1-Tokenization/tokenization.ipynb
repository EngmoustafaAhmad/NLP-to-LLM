```python
import nltk
import string
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
text = "hello , i'm moustafa software engineer app developer"

sentences=nltk.sent_tokenize(text,language="english")
print(sentences)
```

```python
tokens=nltk.word_tokenize(text,language="english")
print(tokens)
```

```python
clean_tokens=[i for i in tokens if i not in string.punctuation]
print(clean_tokens)
```
